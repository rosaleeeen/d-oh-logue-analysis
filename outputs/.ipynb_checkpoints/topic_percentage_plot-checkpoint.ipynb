```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('topic_percentages.csv')
print(df)

characters = df['character']
topics = ['COM', 'CON', 'ECO', 'FAM', 'LEI', 'PER', 'REL', 'WRK']

x = np.arange(len(topics))
width = 0.25

fig, ax = plt.subplots(figsize=(12, 6))

bars1 = ax.bar(x - width, df.loc[df['character'] == 'burns', topics].values[0], 
               width, label='Burns', color='#f0e442')
bars2 = ax.bar(x, df.loc[df['character'] == 'ned', topics].values[0], 
               width, label='Flanders', color='#0072b2')
bars3 = ax.bar(x + width, df.loc[df['character'] == 'skinner', topics].values[0], 
               width, label='Skinner', color='#cc79a7')

ax.set_xlabel('Topics', fontsize=12, fontweight='bold')
ax.set_ylabel('Proportion', fontsize=12, fontweight='bold')
ax.set_title('Topic Distribution by Character', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(topics)
ax.legend()

ax.grid(axis='y', alpha=0.3, linestyle='--')
ax.set_axisbelow(True)

plt.tight_layout()

plt.show()
```
